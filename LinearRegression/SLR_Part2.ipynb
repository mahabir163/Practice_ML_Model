**Import Library**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

**Read data set**

```python
df = pd.read_csv("e:/Users/USER/Desktop/DataAnalytics And gen AI/Data/ML/height-weight.csv")
df
```

```python
#Relaionship between weight and height
plt.scatter(df["Weight"],df["Height"])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
#Problem statement : We want to predict height based on a weight
```

```python
#Divide the data set into x and y
x = df[["Weight"]]
y = df["Height"]
```

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train,x_test
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print("The slope or coedd of model is: ",model.coef_)
```

```python
print("The intercept of the model is: ",model.intercept_)
```

```python
#Predicted height(y) on xtrain = 160 + 17.73 * (xtrain)
#Predicted height(y) on xtest = 160 + 17.73 * (xtest)
```

```python
#Prediction on train data
y_pred = model.predict(x_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred)
```

```python
#Predict for test data
y_pred_test = model.predict(x_test)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred_test)
```

```python
#Performance matrics
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_test)
```
